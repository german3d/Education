```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

from algorithms.neural_network import RBF

%matplotlib inline
```

Fitting randomly generated data with RBF model

```python
np.random.seed(13)

plt.figure(figsize=(12,6));

# true data
X = np.linspace(start=1, stop=10, num=100)
y = np.sin(X) + np.random.normal(loc=0, scale=0.4, size=X.shape)
plt.plot(X, y, ".", label="true data");


# fitted data
rbf = RBF(num_neurons=3, use_kmeans=True, normalize=True)
rbf.fit(X, y)
y_pred = rbf.predict(X)
plt.plot(X, y_pred, ".-", color="darkorange", alpha=0.6,
         label="RBF: 2 neurons, k-means, normalized, MSE={:.2f}".format(mean_squared_error(y,y_pred)))

rbf = RBF(num_neurons=5, use_kmeans=True, normalize=True)
rbf.fit(X, y)
y_pred = rbf.predict(X)
plt.plot(X, y_pred, ".-", color="red", alpha=0.6, 
         label="RBF: 5 neurons, k-means, normalized, MSE={:.2f}".format(mean_squared_error(y,y_pred)))

rbf = RBF(num_neurons=5, use_kmeans=False, normalize=True)
rbf.fit(X, y)
y_pred = rbf.predict(X)
plt.plot(X, y_pred, ".-", color="blue", alpha=0.6, 
         label="RBF: 5 neurons, random, normalized, MSE={:.2f}".format(mean_squared_error(y,y_pred)))

rbf = RBF(num_neurons=5, use_kmeans=False, normalize=False)
rbf.fit(X, y)
y_pred = rbf.predict(X)
plt.plot(X, y_pred, ".-", color="cyan", alpha=0.6, 
         label="RBF: 5 neurons, random, non-normalized, MSE={:.2f}".format(mean_squared_error(y,y_pred)))

plt.title("Fitted RBF with different parameters");
plt.grid();
plt.legend();
```
